```python
# Chrome 웹 드라이버 경로 지정
driver_path = 'C:/RSelenium/chromedriver.exe'  # 본인의 드라이버 경로로 수정

# 웹 드라이버에 서비스 지정
chrome_service = webdriver.chrome.service.Service(driver_path)
chrome_service.start()

```

```python
qbar = tqdm(jachi_link)
count1 = 0

for c in qbar:
    df_minwon = pd.DataFrame({'날짜': [], '자치구': [], '민원내용': []})

    # 웹 드라이버 시작
    driver = webdriver.Chrome(service=chrome_service)
    driver.get(c)
    driver.fullscreen_window()    # 전체화면으로

    # 검색창 선택
    s_box = driver.find_element(By.CSS_SELECTOR,"#header > div.gnbmiddle > div > div.topmenu > ul > li:nth-child(1) > a")
    s_box.click()
    s_box.click()
    time.sleep(.3)

    minwon1 = driver.find_element(By.XPATH,'//*[@id="allmenu"]/div/div[2]/div[1]/ul/li[2]')
    minwon1.click()
    time.sleep(.5)
    driver.fullscreen_window()    # 전체화면으로

    nalja = driver.find_element(By.CSS_SELECTOR,"#dataSetTb > table > tbody > tr:nth-child(1) > td:nth-child(4)").text

    count = 0

    try:
        cd1 = driver.find_element(By.XPATH,'//*[@id="startDate"]')
        cd1.clear()
        cd1.send_keys('20210101')
        cd1.send_keys(Keys.ENTER)
        time.sleep(0.1)

        cd2 = driver.find_element(By.XPATH,'//*[@id="endDate"]')
        cd2.clear()
        cd2.send_keys('20211231')
        cd2.send_keys(Keys.ENTER)
        time.sleep(0.1)

        jh = driver.find_element(By.XPATH,'//*[@id="searchBtn"]')
        jh.click()
        time.sleep(0.1)
        driver.fullscreen_window()
    except:
        pass
    while True:
        if (nalja.startswith('2023')) | (nalja.startswith('2022')):
            id_prev = driver.find_element(By.XPATH, '//*[@id="dataSetTb"]/table/tbody/tr[1]/td[1]')
            next = driver.find_element(By.CSS_SELECTOR,"#navigator > a.navi.navi-arrow.navi-arrow-single-right")
            next.click()
            nalja = driver.find_element(By.CSS_SELECTOR,"#dataSetTb > table > tbody > tr:nth-child(1) > td:nth-child(4)").text
            time.sleep(.3)
            id_next = driver.find_element(By.XPATH, '//*[@id="dataSetTb"]/table/tbody/tr[1]/td[1]')
            if id_prev == id_next:
                driver.close()
                break

        # 공개상담민원 화면
        # previous_url = driver.current_url
        elif nalja.startswith('2021'): 
            j = 1
            while j < 20:
                pbar = tqdm(range(j, j + 2))
                for i in pbar:
                    if i > 1:
                        try:
                            try:
                                duiro = driver.find_element(By.XPATH,'/html/body/main/div/div[3]/a')
                            except:
                                duiro = driver.find_element(By.CSS_SELECTOR,'body > main > div > div.btn-group.align-right > a')
                            duiro.click()
                            driver.fullscreen_window()    # 전체화면으로
                            time.sleep(1)
                        except:
                            pass

                        driver.fullscreen_window()    # 전체화면으로
                        # minwon = driver.find_element(By.CSS_SELECTOR,f'#dataSetTb > table > tbody > tr:nth-child({i}) > td.td-list')
                        minwon = driver.find_element(By.XPATH,f'//*[@id="dataSetTb"]/table/tbody/tr[{i}]/td[2]/a')
                        
                        minwon.click()
                        driver.fullscreen_window()    # 전체화면으로
                        time.sleep(1)

                        try:
                            # 민원 내용 크롤링
                            # nalja = driver.find_element(By.CSS_SELECTOR,"body > main > div > fieldset > table > tbody > tr:nth-child(2) > td:nth-child(4)").text
                            nalja = driver.find_element(By.XPATH,"/html/body/main/div/fieldset/table/tbody/tr[2]/td[2]").text
                            nalja = int(re.sub('[^0-9]', '', nalja))

                            location = jachi[count1]

                            contents = driver.find_element(By.CSS_SELECTOR,"body > main > div > fieldset > table > tbody > tr:nth-child(4) > td").text
                        except:
                            pass

                    else:
                        driver.fullscreen_window()    # 전체화면으로
                        time.sleep(.5)
                        minwon = driver.find_element(By.XPATH,f'//*[@id="dataSetTb"]/table/tbody/tr[{i}]/td[2]/a')
                        minwon.click()
                        driver.fullscreen_window()    # 전체화면으로
                        time.sleep(1)

                        # 민원 내용 크롤링
                        nalja = driver.find_element(By.XPATH,"/html/body/main/div/fieldset/table/tbody/tr[2]/td[2]").text
                        nalja = int(re.sub('[^0-9]', '', nalja))

                        location = jachi[count1]

                        contents = driver.find_element(By.CSS_SELECTOR,"body > main > div > fieldset > table > tbody > tr:nth-child(4) > td").text

                    df_minwon.loc[count, :] = [nalja, location, contents]
                    count += 1
                
                j += 2
                # nalja = driver.find_element(By.CSS_SELECTOR,"#dataSetTb > table > tbody > tr:nth-child(1) > td:nth-child(4)").text
                if j > 19:
                    try:
                        duiro = driver.find_element(By.XPATH,'/html/body/main/div/div[3]/a')
                    except:
                        duiro = driver.find_element(By.CSS_SELECTOR,'body > main > div > div.btn-group.align-right > a')
                    duiro.click()
                    time.sleep(.3)
                    
                    next = driver.find_element(By.CSS_SELECTOR,"#navigator > a.navi.navi-arrow.navi-arrow-single-right")
                    next.click()
                    time.sleep(.3)
                    nalja = driver.find_element(By.CSS_SELECTOR,"#dataSetTb > table > tbody > tr:nth-child(1) > td:nth-child(4)").text
        elif nalja.startswith('2020'):
            driver.close()
            break


    df_minwon['날짜'] = round(df_minwon['날짜'] / (10**6)).astype(int)
    print(df_minwon.tail())
    df_minwon.to_csv(f'{jachi[count1]} 민원 데이터.csv', index=False, encoding='CP949')

    print(f'{jachi[count1]} 민원 저장 완료')
    count1 += 1

qbar.close() ## 반복이 끝나고 close 사용 강력 권장

```
